## Imports

```python
import torch
import torch.nn as nn
import numpy as np
```

## Model

```python
class Perceptron(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=52, out_features=256),
            nn.ReLU(),
            nn.Linear(in_features=256, out_features=128),
            nn.ReLU(),
            nn.Linear(in_features=128, out_features=10),
            nn.Softmax(dim=1)
        )


    def forward(self, x):
        return self.model(x)
```
